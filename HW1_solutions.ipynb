## Question 4

Let's do some plotting to continue our EDA process.

**Part (a)**: Make a plot to assess the distribution of total_personal_income in the data. Describe the distribution using your plot.

```python
import seaborn as sns
sns.set_theme()

sns.histplot(data=df, x="total_personal_income")
```

The distribution is very right-skewed, and has a sharp cut-off at zero on the left.

**Part (b)**: Make a plot to assess whether or not there is a relationship between age and total_personal_income in the data. Use your plot to assess whether or not you think such a relationship exists.

```python
sns.scatterplot(data=df, x="age", y="total_personal_income")
```

People in age ranges between 35-75 seem to have the ability to make more money, although on average there doesn't seem to be a very significant relationship.

**Part (c)**: Make a plot to see if there is a relationship between sex and total personal income, and again interpret your plot.

```python
# Violin plot or boxplot would be reasonable

sns.violinplot(data=df, x="sex", y="total_personal_income")
```

The distributions of total personal income are definitely different for male and female, with males having a much longer right tail (i.e. many more males in the highest income ranges).

**Part (d)**: Make a plot comparing age and total_personal_income as in part (a), but this time stratify by sex as well, in order to see whether or not sex impacts the relationship between age and income. Interpret your plot.

```python
# Natural choice is colored scatterplot
sns.scatterplot(data=df, x="age", y="total_personal_income", hue="sex")
```

It's clear again that the individuals in the highest income ranges are males, but beyond this, the relationship between age and income seems to be about the same for both sexes.
